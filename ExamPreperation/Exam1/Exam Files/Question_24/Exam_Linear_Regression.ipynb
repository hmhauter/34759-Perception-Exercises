````python

x = np.loadtxt('/home/hhauter/Documents/S23/Perception/34759-Perception-Exercises/ExamPreperation/Exam Files/Question_24/lr_x.txt')

target = np.loadtxt('/home/hhauter/Documents/S23/Perception/34759-Perception-Exercises/ExamPreperation/Exam Files/Question_24/lr_y.txt')

````

````python
x_reshaped = x.reshape(-1, 1) #fit needs data in this shape
# Initialize and fit model
lm = LinearRegression()
model = lm.fit(x_reshaped, target)
````

````python
resulting_y = model.predict(np.array([0]).reshape(-1, 1))
print(resulting_y)
````

````python
resulting_y2 = model.predict(np.array([1]).reshape(-1, 1))
print(resulting_y2)
````

````python
print(resulting_y2-resulting_y)
````
