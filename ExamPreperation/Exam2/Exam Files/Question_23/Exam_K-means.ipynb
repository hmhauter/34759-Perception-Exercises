````python
import matplotlib.pyplot as plt  
from matplotlib import style 
from sklearn.cluster import KMeans 
# from sklearn.datasets.samples_generator import make_blobs 
from sklearn.decomposition import PCA
import numpy as np

# Loading the data
X = np.loadtxt('clusters.txt', dtype=int)
````

# Sub-Question A - K-means
Given the dataset 'clusters.txt' that has been read into the variable X, find how many clusters are optimal for this dataset using the elbow method.


Recall you call Kmeans:
```{python}
km = KMeans(n_clusters = n)
km.fit(X)
```
Documentation:
https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

````python
wcss = []

# Compute the within-cluster sum of squares for values of k from 1 to 10
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) 

# Plot the within-cluster sum of squares against the number of clusters
plt.plot(range(1, 15), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-Cluster Sum of Squares')
plt.show()
````
