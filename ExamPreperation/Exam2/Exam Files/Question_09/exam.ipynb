```python
bgr_img = cv2.imread('books.png')
b,g,r = cv2.split(bgr_img)       # get b,g,r
image = cv2.merge([r,g,b])
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap = 'gray')
```

```python
edged = cv2.Canny(gray, 100, 200)
plt.imshow(edged, cmap='gray')
```

```python
# Apply HoughLines (cv2.HoughLines) with the parameters "rho=1",
# "theta=0.0017", "threshold=200".

lines = cv2.HoughLines(edged, 1, 0.0017, 200) 

print(len(lines))
```

```python
# Apply Harris corner (cv2.cornerHarris) to the image with the parameters
# "blockSize=2", "ksize=3", "k=0.04".
counter = 0
sol = cv2.cornerHarris(gray,2,3,0.04)
for i in range(sol.shape[0]):
    for j in range(sol.shape[1]):
        if(sol[i][j] > 0.01):
            counter += 1

print(counter)
```
