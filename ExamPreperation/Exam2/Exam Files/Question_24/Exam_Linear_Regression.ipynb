````python
x = np.loadtxt('lr_x.txt', dtype=float)
y = np.loadtxt('lr_y.txt', dtype=float)
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'ro')   # scatter plot showing actual data
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

````python
resulting_y = model.predict(np.array([0]).reshape(-1,1))
print(resulting_y)

resulting_y = model.predict(np.array([1]).reshape(-1,1))
print(resulting_y)
````
